## Red Wine Quality Prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("winequality-red.csv")
```

```python
df.head()
```

```python
df.isnull().sum() 
```

```python
## no data is missing we can dirctly train the model
```

```python
## let assign x and y values

x=df.iloc[:,:-1]  ## taking all independent feature except quality
y=df.iloc[:,11]
```

```python
y.head() ## dependent varible is perfectly selected
```

```python
x.head()  ## except quality all other independent variables are selected
```

## now we need to split our data into validation and train set


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=50)
```

## now import Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_prediction=model.predict(x_test)
print(pd.DataFrame(y_prediction))
```


##  We need to check where our predictions are correct

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_prediction)
```

```python
print(score)
```
